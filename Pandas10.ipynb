## Quiz ตัดตารางออกมาเฉพาะปี 2021 ```announce_date``` ในปี 2021 

**Hint**
*   วนลูปหา index ของปี 2021
*   ตัดตารางมาเฉพาะ ปี 2021



````python
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021': # อยากได้แค่ปี 2021 ใช้การวนลูปเอา
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````

### HW  กลุ่ม 2 || เลือกข้อมูลมาเฉพาะวัน annouce_date เป็นวันที่ 1-20

````python
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[1] <= '20':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF]
````

#Data Preprocessing

##Data1

ตัดคอลัมน์ที่ไม่ต้องการออกไปใน Data1 (เลือกเฉพาะคอลัมน์ที่ต้องการ)



```python
import numpy as np
```

```python
data1_1 = data1[['user','RowKey','date','ov','phone','site','time','urine','PartitionKey','title','age']]   # ชี้ข้อมูลคอลัมน์ที่ต้องการ
data1_1
```

ตรวจสอบค่า missing

```python
data1_1.isnull().any() # ตรวจสอบหาค่า missing
```

```python
data1_2 = data1_1.dropna()  # dropna ลบค่า missing ออก
data1_2
```

เช็คประเภทของข้อมูลแต่ละคอลัมน์

```python
data1_2.dtypes
```

### user

คอลัมน์ user เลือกเฉพาะแถวที่เป็นตัวเลข 5 หลัก และแปลง object ให้เป็น integer

```python
set(data1_2['user'])
```

```python
cols = ['user'] #กำหนดคอลัมน์สำหรับเช็คตัวเลข

mask = data1_2[cols].apply(lambda x: pd.to_numeric(x, errors='coerce')).notnull().all(axis=1) #กรองเฉพาะแถวที่เป็นตัวเลข
data1_3 = data1_2[mask]

data1_3[cols] = data1_3[cols].astype(int) #แปลงจาก string เป็น integer
data1_3
```

```python
data1_4 = data1_3[(data1_3['user'] > 9999) & (data1_3['user']< 100000)] #เลือกมาเฉพาะแถวที่มีตัวเลข 5 หลัก
data1_4
```

```python
set(data1_4['user'])
```

###RowKey

```python
data1_4['RowKey'] = data1_4['RowKey'].astype('string')
data1_4['RowKey']
```

###date

แปลงคอลัมน์ date จาก object ให้เป็น datetime
